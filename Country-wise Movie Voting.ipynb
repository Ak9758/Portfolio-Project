```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\Admin\Desktop\DA Projects\Top IMDB Rated TV Shows.csv')
df
```

```python
#showing dataframes only where vote_average>8
df = df[df['vote_average'] >= 8]
df
```

```python
countrywise= df.groupby('origin_country')
countrywise.first()
```

```python
countrywise.first().sort_values(by=['vote_average'],ascending=False)
```
